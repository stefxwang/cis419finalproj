```python
import json
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer
import datetime
import numpy as np
```

```python
data = json.load(open('modified_Data.json'))["data"]
```

```python
n = Normalizer()
dayofweek = list()
createdTime = list()
likes = list()
dayofweekCount = np.zeros(7)
dayofweekLikesCount = np.zeros(7)
```

```python
for i in range(len(data)):
    dayofweek.append(data[i]['dayofweek'])
    dayofweekCount[data[i]['dayofweek']] += 1
    dayofweekLikesCount[data[i]['dayofweek']] += data[i]['likes']
    '''ordinal time is 1 per day'''
    createdTime.append(datetime.date.fromordinal(data[i]['created_time']))
    likes.append(data[i]['likes'])
```

```python
plt.figure(1, figsize=(20,10))
plt.scatter(createdTime, likes, 0.5)

plt.figure(2, figsize=(20,10))
plt.scatter(dayofweek, likes, 0.5)

plt.figure(3, figsize=(20,10))
plt.hist(likes)

plt.figure(4, figsize=(20,10))
plt.bar(np.arange(7), dayofweekCount)

plt.figure(5, figsize=(20,10))
plt.bar(np.arange(7), dayofweekLikesCount)

plt.figure(6, figsize=(20,10))
plt.hist(createdTime)

plt.figure(7, figsize=(20,10))
plt.hist(createdTime, 80, [datetime.date(2016, 1, 1), datetime.date(2017, 11, 22)])
```
